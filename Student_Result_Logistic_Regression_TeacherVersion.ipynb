```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv("Student_Result.csv")
df.head()
```

```python
#EDA
df.shape
```

```python
# Find missing values
df.isnull().sum()
```

```python
df.info()
```

```python
#Extract Input and Output columns(Independed and Depended variables)
X=df.drop('Pass_Exam',axis='columns')
y=df.Pass_Exam
```

```python
X.shape
```

```python
y.shape
```

```python
model=LogisticRegression()
```

```python
model.fit(X,y)
```

```python
model.predict([[8.5]])
```

```python
model.score(X,y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
len(y_train)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
model.score(X_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
